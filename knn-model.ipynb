```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
trainData = pd.read_csv('titantic_data/train.csv')
print(np.shape(trainData))
trainData.head(10)
```

```python
#store passengerIds
passId = trainData.values[:,0]

#store sex
sex = trainData.values[:,4]

#store passenger names
passName = trainData.values[:,3]

#store ticket
ticket = trainData.values[:,8]

#store cabin
cabin = trainData.values[:,10]
```

```python
#dropping columns... passengerId and Name seem irrelvant and not sure what to do with ticket and cabin
trainData = trainData.drop(['PassengerId','Name','Ticket','Cabin'], axis=1)
trainData.head(10)
```

```python
#onehot encoding for sex and embarked
encodeTrainData = pd.get_dummies(trainData, columns=['Sex','Embarked']).astype(float)
encodeTrainData.head(10)
```

```python
#for any NaN values in data change replace with median to not skew data
medians = encodeTrainData.median()
newTrainData = encodeTrainData.fillna(medians)
newTrainData.head(10)
```

```python
newTrainData.to_csv('newTrainData.csv', index=False)
```

```python
#split dataset
x = newTrainData.iloc[:, 1:]
y = newTrainData.iloc[:, 0]

#random_state is for random seed, test_size is percentage of training data we want to use for validation data
x_train, x_valid, y_train, y_valid = train_test_split(x, y, random_state=0, test_size=0.2)
```

```python
#feature scaling for standardizing dataset
sc = StandardScaler()

#creating scaler on x_train variable
x_train = sc.fit_transform(x_train)

#this isnt part of the scaler
x_valid = sc.transform(x_valid)
```

```python
#define the model -> K = 10, p=2(survived or dead)
classifier = KNeighborsClassifier(n_neighbors=20, p=2, metric='euclidean')
```

```python
#fit model
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_valid)
y_pred
```

```python
cm = confusion_matrix(y_valid, y_pred)
cm
```

```python
print(f1_score(y_valid, y_pred))
```

```python
print(accuracy_score(y_valid, y_pred))
```
