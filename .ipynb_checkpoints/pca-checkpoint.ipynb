```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
trainData = pd.read_csv('newTrainData.csv')
trainData.head()
```

```python
survival = trainData['Survived'].copy()

trainData = trainData.drop(['Survived'], axis=1)
trainData.head()
```

```python
# # MinMaxScaler transforms features by scaling featured from 0 - 1
# from sklearn.preprocessing import MinMaxScaler

# z-score normalization
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(trainData)
```

```python
scData = sc.transform(trainData)
```

```python
scData
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=5)
```

```python
pca.fit(scData)
```

```python
xPCA = pca.transform(scData)
```

```python
scData.shape
```

```python
xPCA.shape
```

```python
colorMap = {0: 'red', 1: 'blue'}
plt.figure(figsize=(8,6))

for label in [0, 1]:
    # Filter data by label
    indices = survival == label
    plt.scatter(xPCA[indices, 0], xPCA[indices, 1], 
                c=colorMap[label], label=label, edgecolor='k')
    
plt.legend(title="Survival")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report, accuracy_score
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(xPCA, survival, test_size = 0.2, random_state=0)
```

```python
knn = KNeighborsClassifier(n_neighbors=20, p=2, metric='euclidean')
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_valid)
```

```python
print("Accuracy:", accuracy_score(y_valid, y_pred))
print("\nClassification Report:\n", classification_report(y_valid, y_pred))
```

```python
testData = pd.read_csv('titantic_data/test.csv')
testData.head()
```

```python
passID = testData['PassengerId'].copy()
```

```python
testData = testData.drop(['PassengerId','Name','Ticket','Cabin'], axis=1)
testData.head(20)
```

```python
encodeTestData = pd.get_dummies(testData, columns=['Sex','Embarked']).astype(float)
encodeTestData.head(20)
```

```python
medians = encodeTestData.median()
newTestData = encodeTestData.fillna(medians)
newTestData.head(20)
```

```python
sc.fit(newTestData)
```

```python
scTestData = sc.transform(newTestData)
```

```python
scTestData
```

```python
test_pca = PCA(n_components=5)
test_pca.fit(scTestData)
```

```python
x_test_pca = test_pca.transform(scTestData)
```

```python
x_test_pca.shape
```

```python
test_pred = knn.predict(x_test_pca).astype(int)
```

```python
test_pred
```

```python
output = pd.DataFrame({'PassengerId': passID, 'Survived': test_pred})
output.head(20)
```
