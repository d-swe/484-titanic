```python
import numpy as np
import pandas as pd

trainData = pd.read_csv('titantic_data/train.csv')
print(np.shape(trainData))
trainData.head(10)
```

```python
labels = trainData.values[:,1]

#store passenger names
passName = trainData.values[:,3]

#drop passenger names
newTrainData = trainData.drop('Name', axis=1)

#store sex
sex = trainData.values[:,4]

#create a sexy dataframe
sexyDF = pd.DataFrame(sex, columns=['Sex'])

#one-hot encode the sexy dataframe
oheSexy = pd.get_dummies(sexyDF['Sex'], prefix='Sex')
oheSexy = oheSexy.astype(int)

#drop sex
newTrainData = newTrainData.drop('Sex', axis=1)

#store ticket
ticket = trainData.values[:,8]

#store cabin
cabin = trainData.values[:,10]

#store embarked
embark = trainData.values[:,11]

#create embark dataframe
embarkDF = pd.DataFrame(embark, columns=['Embarked'])

#one-hot encode embark dataframe
oheEmbark = pd.get_dummies(embarkDF['Embarked'], prefix='Embarked')
oheEmbark = oheEmbark.astype(int)
oheEmbark.head()

#drop survived, ticket, cabin, embarked. we can add them back just not sure how rn
newTrainData = newTrainData.drop(['Survived','Ticket','Cabin','Embarked'], axis=1)
print(f"newTrainData Shape: {np.shape(newTrainData)}")

#add one-hot encodings for sex and embarked
newTrainData = pd.concat([newTrainData, oheSexy, oheEmbark], axis=1)
newTrainData.head(10)
```

```python
#creating new validation data, made into list for ez append
validData = []
validLabels = []
#convert it back to ndarray
trainDataArr = newTrainData.values

#need to replace NaN in age column
median_age = np.nanmedian(trainDataArr[:,2])
#where ever there is a NaN in age column replace with median_age
trainDataArr[:,2] = np.where(np.isnan(trainDataArr[:,2]), median_age, trainDataArr[:,2])
```

```python
#using 80 to 20 ratio -> training data to validation data
s = (int)(len(newTrainData) * .2)

#randPick will be a set of random integers
randPick = set()
randPick = np.random.randint(0, len(newTrainData), size=s)
randPick.sort()

for r in randPick:
    validData.append(trainDataArr[r])
    validLabels.append(labels[r])
    
#conver it back to ndarray
validData = np.array(validData)

#checking structures of valid and train
print(f"ValidData shape: {np.shape(validData)}")
print(f"validLabels shape: {np.shape(validLabels)}")
ndf = pd.DataFrame(trainDataArr)
ndf.head(10)
```

```python
trainDataArr = list(trainDataArr)
labels = list(labels)
randPick = randPick[::-1]

#remove the rows picked for validation dataset from training dataset
for r in randPick:
    trainDataArr.pop(r)
    labels.pop(r)
```

```python
#separate labels for training and valid datas
train_labels = np.array(labels)
print(f"Train labels shape: {np.shape(train_labels)}")

valid_labels = np.array(validLabels)
print(f"Valid labels shape: {np.shape(valid_labels)}")
```

```python
np.savetxt("train_labels.txt", train_labels, delimiter=",", fmt="%.5f")
np.savetxt("valid_labels.txt", valid_labels, delimiter=",", fmt="%.5f")
np.savetxt("trainData.txt", trainDataArr, delimiter=",", fmt="%.5f")
np.savetxt("validData.txt", validData, delimiter=",", fmt="%.5f")
```
